```python
b = pd.Series([1,2,3,4])
```

```python
print(b)
```

```python
print("Dimension - ",b.ndim)
print("Datatype - ",b.dtype)
print("Shape -" ,b.shape)
print("Values -" ,b.values)
print("Index -" ,b.index)
print("Size -" ,b.size)
print("is Empty-",b.empty)
print("Converting to lsit -" ,b.tolist())
```

```python
b= pd.Series([1,22,32,3,3,3,4,15])
print("Unique Elements" ,b.unique()) # it prints which are not duplicated
print("Total Unique Elements ",b.nunique())
print("Value Counts",b.value_counts())
```
