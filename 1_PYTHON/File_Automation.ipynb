```python
import os
def file_organize(path):
     #print(path)
     extensions=[]
    
     files=os.listdir(path)
     for i in files:
         x=os.path.join(path,os.path.splitext(i)[1][1:])
         #print(x)
         if x not in extensions:
             extensions.append(x)
     for folder  in extensions:
         try:
             os.mkdir(folder)
         except:
             pass
         for file in files:
             if(os.path.split(folder)[1]==os.path.splitext(file)[1][1:]):
                 os.rename(os.path.join(path,file),os.path.join(folder,file))

             
mypath="C:\\Users\\thula\\OneDrive\\Desktop\\Automation"
file_organize(mypath)
```
