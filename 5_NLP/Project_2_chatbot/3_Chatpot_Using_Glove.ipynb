```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import gensim.downloader as api


glove = api.load("glove-wiki-gigaword-50")


answers = [
    "hello! i am your chatbot.",
    "i can help you with python, machine learning, and nlp basics.",
    "python is a programming language.",
    "machine learning is teaching computers to learn patterns.",
    "nlp stands for natural language processing.",
    "goodbye and have a nice day!"
]


def sentence_vector(text):
    words = text.lower().split()
    vecs = [glove[w] for w in words if w in glove]
    if not vecs:
        return np.zeros(glove.vector_size)
    return np.mean(vecs, axis=0)


answer = np.array([sentence_vector(a) for a in answers])

def chatbot_response(user_input):
    user_vec = sentence_vector(user_input).reshape(1, -1)
    sims = cosine_similarity(user_vec, answer)[0]

    best_idx = np.argmax(sims)
    best_score = sims[best_idx]

    if best_score > 0.2:
        return answers[best_idx]
    else:
        return "I am sorry, I donâ€™t understand that."

while True:
    user_input = input("You: ")
    if user_input.lower() in ["bye", "exit", "quit"]:
        print("Chatbot: Goodbye! ðŸ‘‹")
        break
    print("Chatbot:", chatbot_response(user_input))
```
