```python
documents = [
    "My students love Python, python python python python python is great",
    "Python is great",
    "My students like Python"
]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(documents)
```

```python
vectorizer.vocabulary_
```
